```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./datasets/data.csv')
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=30)
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = knn.predict(X_test)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
```

```python
# finding all accuracy scores by changing the k-values.
import math

# Method-1: Heuristic
k1 = int(math.sqrt(df.shape[0]))
knn2 = KNeighborsClassifier(n_neighbors=k1)
knn2.fit(X_train, y_train)
y_pred2 = knn2.predict(X_test)
print(f"Accuracy score: {accuracy_score(y_test, y_pred2)}")
```

```python
# Method-2: Experimental

scores = []
for i in range(1, 200):  # taken experimentally, you can take till any value
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred3 = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred3))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(scores)
plt.grid(True)
plt.show()
```
